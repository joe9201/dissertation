```python
import os
import numpy as np
import pandas as pd
import networkx as nx
from causallearn. search.FCMBased import lingam
from plotting_utils import plot_and_save_graph
from data_preparation import load_and_prepare_student_data, load_and_prepare_adult_data
from true_graph import create_true_graph_student, create_true_graph_student_small, create_true_graph_adult, create_true_graph_adult_small
from evaluation import evaluate_graph
import matplotlib.pyplot as plt
```

```python
def run_lingam_algorithm(data, labels):
    model_lingam = lingam.ICALiNGAM(max_iter=2000, indep_test)
    model_lingam.fit(data)
    
    adjacency_matrix = model_lingam.adjacency_matrix_

    # Create NetworkX graph for evaluation
    lingam_graph = nx.DiGraph(adjacency_matrix)

    # Relabel nodes using the provided labels
    mapping = {i: labels[i] for i in range(len(labels))}
    lingam_graph = nx.relabel_nodes(lingam_graph, mapping)

    plot_and_save_graph(lingam_graph, labels, 'lingam_graph.png')

    return lingam_graph
```

```python
def load_dataset(dataset):
    if dataset == 'adult':
        data_file = 'data/adult_cleaned.csv'
        df_encoded, labels, data = load_and_prepare_adult_data(data_file)
        true_graph = create_true_graph_adult()

    elif dataset == 'adult_small':
        data_file = 'data/smaller_adult_dataset.csv'
        df_encoded = pd.read_csv(data_file)
        labels = df_encoded.columns.tolist()
        data = df_encoded.to_numpy()
        true_graph = create_true_graph_adult_small()

    elif dataset == 'student':
        data_file = 'data/student-por_raw.csv'
        df_encoded, labels, data = load_and_prepare_student_data(data_file)
        true_graph = create_true_graph_student()

    elif dataset == 'student_small':
        data_file = 'data/smaller_student_dataset.csv'
        df_encoded = pd.read_csv(data_file)
        labels = df_encoded.columns.tolist()
        data = df_encoded.to_numpy()
        true_graph = create_true_graph_student_small()

    else:
        raise ValueError("Invalid dataset specified. Choose 'adult', 'adult_small', 'student', or 'student_small'.")

    return df_encoded, labels, data, true_graph
```

```python
def main(dataset):
    # Load the selected dataset
    df_encoded, labels, data, true_graph = load_dataset(dataset)
    
    print(f"Running LiNGAM on {dataset} dataset.")
    
    # Run LiNGAM algorithm
    lingam_graph = run_lingam_algorithm(data, labels)
    
    print("LiNGAM Algorithm graph created.")
    
    # Evaluate the estimated graph
    shd, recall, precision = evaluate_graph(lingam_graph, true_graph)
    print(f"Structural Hamming Distance (SHD): {shd}")
    print(f"Recall: {recall}")
    print(f"Precision: {precision}")
```

```python
dataset = 'student_small'  # Change to 'adult', 'adult_small', 'student', or 'student_small'
main(dataset)
```
